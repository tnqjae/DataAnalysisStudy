```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X,Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots()

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1, Y_1)
ax.set(title ='Cos Graph', xlabel = 'X', ylabel = 'Y')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(2,2,1)
ax_2 = fig.add_subplot(2,2,2)
ax_3 = fig.add_subplot(2,2,3)
ax_4 = fig.add_subplot(2,2,4)

ax_1.plot(X_1, Y_1, c="b")
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

```python
#축 여러 개
#그림 객체에서 add_subplot()사용
#plot객체에서 subplots() 사용하여 축 객체 공간 확보
fig, ax = plt.subplots(nrows=3, ncols=4)
print(ax)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100) #범위 지정(시작값, 종료값, 데이터 개수)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, y_1)
ax[0,1].plot(x, y_2)
ax[1,0].plot(x, y_3)
ax[1,1].plot(x, y_4)
```

```python
#좌표계는 각각 독립적으로 생각해야함. 하나의 판이 아니라 여러개의 판을 덧대는 형식
ax1 = plt.subplot(321)
plt.plot(x, y_1)
ax2 = plt.subplot(322)
ax3 = plt.subplot(312)
ax4 = plt.subplot(325)
ax5 = plt.subplot(326)
```

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value + 10 for value in X]

plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")

plt.show()
```

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c="b")
ax_1.set_title("Figure 1")
ax_2.plot(X_2, Y_2, c="g")
ax_2.set_title("Figure 2")

plt.show()
```
